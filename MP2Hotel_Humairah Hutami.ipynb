```python
# Mengatasi Missing Values

## Nilai kosong pada kolom "agent" dapat dianggap bahwa pemesanan dilakukan tanpa melalui agen.
## Nilai kosong pada kolom "company" dapat dianggap bahwa pemesanan tersebut dilakukan untuk personal/pribadi.
na_replace = {'children':0.0, 'city':'Unknown', 'agent':0, 'company':0}
df_clean = df.fillna(na_replace)

## Nilai 'Undefined' pada kolom "meal" sama dengan "SC" (SC = Self-Catering).
df_clean['meal'].replace('Undefined', 'Self Catering', inplace = True)

## Membuang baris yang memiliki jumlah tamu 0 adults, 0 children, & 0 babies
no_guest = list(df_clean.loc[df_clean['adults'] +
                             df_clean['children'] +
                             df_clean['babies']==0].index)
df_clean.drop(df_clean.index[no_guest], inplace = True)
```
